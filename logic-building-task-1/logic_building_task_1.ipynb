```python
#1.user login check
print("----- user login check-----")

username = "admin"
password = "1234"

input_username = "admin"
input_password = "1234"

if input_username == username and input_password == password:
    print("login successful")
else:
    print("invalid credentials")
```

```python
#2.pass / fail analyzer
print("\n----- pass / fail analyzer -----")

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("total passed students:", pass_count)
print("total failed students:", fail_count)
```

```python
#3.simple data cleaner
print("\n----- simple data cleaner -----")

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned = name.strip().lower()
    cleaned_names.append(cleaned)

print("cleaned names:", cleaned_names)
```

```python
#4.message length analyzer
print("\n----- message length analyzer -----")

messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print(f"Message: '{message}' | Length: {length}")

    if length > 10:
        print("message exceeds 10 characters")
```

```python
#5.error message detector
print("\n----- error message detector -----")

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = logs.count("ERROR")

print("total ERROR entries:", error_count)
```
