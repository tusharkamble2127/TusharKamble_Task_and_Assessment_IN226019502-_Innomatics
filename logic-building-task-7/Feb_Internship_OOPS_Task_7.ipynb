##Problem 1: Smart Light Controller

This class simulates a smart light system that can be turned ON or OFF.
It demonstrates basic object-oriented concepts like class, constructor, and methods.

```python
class SmartLight:

    # Constructor to initialize light name and default status
    def __init__(self, name):
        self.name = name
        self.status = "OFF"   # Default state

    # Method to turn light ON
    def turn_on(self):
        self.status = "ON"
        print(f"{self.name} is {self.status}")

    # Method to turn light OFF
    def turn_off(self):
        self.status = "OFF"
        print(f"{self.name} is {self.status}")


# Sample Input
light = SmartLight("Bedroom Light")

# Perform action
light.turn_on()
```

##Problem 2: Employee ID Card System

This class stores employee details such as name, ID, and department.
It displays formatted ID card information using object methods.

```python
class Employee:

    # Constructor to store employee details
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    # Method to display ID card details
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Sample Input
emp = Employee("Rahul", "EMP102", "IT")
emp.display_id_card()
```

##Problem 3: Mobile Contact Record

This class stores contact name and phone number.
It simulates saving and displaying contact information in a mobile application.

```python
class Contact:

    # Constructor to store contact details
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    # Method to display contact information
    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Sample Input
contact1 = Contact("Anita", "9876543210")
contact1.display_contact()
```

##Problem 4: Product Price Tag Generator

This class stores product name and price and generates a formatted price tag.
It represents a simple retail billing system model.

```python
class Product:

    # Constructor to store product details
    def __init__(self, name, price):
        self.name = name
        self.price = price

    # Method to print formatted price tag
    def display_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


# Sample Input
product1 = Product("Headphones", 2499)
product1.display_price_tag()
```

Problem 5: Movie Rating Display System

This class stores movie name and rating and displays formatted movie details.
It simulates how streaming platforms show movie information.

```python
class Movie:

    # Constructor to store movie details
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    # Method to display movie rating
    def display_movie(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


# Sample Input
movie1 = Movie("Inception", 4.8)
movie1.display_movie()
```

##Problem 6: Delivery Address Manager

This class stores customer name and delivery address.
It prints formatted delivery details for courier/logistics systems.

```python
class Delivery:

    # Constructor to store delivery details
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    # Method to print delivery details
    def display_delivery(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# Sample Input
delivery1 = Delivery("Suman", "Hyderabad")
delivery1.display_delivery()
```
