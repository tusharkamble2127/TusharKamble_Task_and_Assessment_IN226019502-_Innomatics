#Problem 1: Employee Performance Bonus Eligibility


```python
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}

# find highest score
max_score = max(employees.values())

# find all employees with highest score
top_performers = [name for name, score in employees.items() if score == max_score]

print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {max_score})")
```

#Problem 2: Search Query Keyword Analysis

```python
import string

query = "Buy mobile phone buy phone online"

# convert to lowercase
query = query.lower()

# remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))

words = query.split()

frequency = {}

for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# display words searched more than once
result = {k: v for k, v in frequency.items() if v > 1}

print(result)
```

#Problem 3: Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# List to store valid sensor readings (even numbers)
valid_readings = []

#Loop through the sensor readings with index
for index, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

#Problem 4: Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Dictionary to store count of each domain
domain_count = {}

#Extract domain from each email and count occurrences
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

#Calculate total number of emails
total = len(emails)

#Calculate and display percentage usage of each domain
for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(f"{domain}: {percentage:.0f}%")
```

# Problem 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
#calculating average
avg=sum(sales)/len(sales)
#logic
for i in range(len(sales)):
  if sales[i]>avg:
    print("Day", i+1, ":", sales[i])
```

#Problem 6: Duplicate User ID Detection


```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Dictionary to store frequency of each user ID
frequency = {}

#Count how many times each user ID appears
for user in user_ids:
    frequency[user] = frequency.get(user, 0) + 1

# Display only duplicate user IDs (count > 1)
for user, count in frequency.items():
    if count > 1:
        print(f"{user} → {count} times")
```
